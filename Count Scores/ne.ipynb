```python
import pandas as pd

data = pd.read_csv('VoteLists.csv', encoding='UTF-8', encoding_errors='replace', header=0, index_col=0)
data
```

```python
participants = data.columns

tables = []
for part in participants:
    col = data.iloc[1:7][part]
    col.name = data.iloc[0].name
    tables.append(col)
    
    col = data.iloc[8:14][part]
    col.name = data.iloc[7].name
    tables.append(col)
    
    col = data.iloc[15:20][part]
    col.name = data.iloc[14].name
    tables.append(col)
tables
```

```python
from pprint import pprint
```

```python
scores = {
    'Эстетичность:': {},
    'Культурность:': {},
    'Всратость:': {},
    'Общий': {},
}

scores['Общий'] = {}

for table in tables:
    category = table.name
    
    for rank, title in table.items():
        cur_score = scores[category].get(title, 0)
        scores[category][title] = cur_score + (7 - int(rank))
        
        cur_score = scores['Общий'].get(title, 0)
        scores['Общий'][title] = cur_score + (7 - int(rank))

pprint(scores)
```

```python
s_func = lambda x: -x[1]

sorting = lambda dick: sorted(dick.items(), key=s_func)

top_v = [f'{ind+1}. {elem[0]}: {elem[1]}' for ind, elem in enumerate(sorting(scores['Эстетичность:']))]
top_k = [f'{ind+1}. {elem[0]}: {elem[1]}' for ind, elem in enumerate(sorting(scores['Культурность:']))]
top_o = [f'{ind+1}. {elem[0]}: {elem[1]}' for ind, elem in enumerate(sorting(scores['Всратость:']))]
top_e = [f'{ind+1}. {elem[0]}: {elem[1]}' for ind, elem in enumerate(sorting(scores['Общий']))]
```

```python
pprint(top_v)
pprint(top_k)
pprint(top_o)
pprint(top_e)
```
